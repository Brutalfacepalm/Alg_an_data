```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
def eval_sgd_model(X, y, iterations, qty_in_batch, alpha=1e-4):
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    n_batch = n // qty_in_batch
    if n % qty_in_batch != 0:
        n_batch += 1
    for i in range(1, iterations+1):
        for b in range(n_batch):
            start_ = qty_in_batch*b
            end_ = qty_in_batch*(b+1)

            # print(b, n_batch, start_, end_)

            X_tmp = X[:, start_ : end_]
            y_tmp = y[start_ : end_]
            y_pred_tmp = np.dot(W, X_tmp)
            err = calc_mse(y_tmp, y_pred_tmp)
            W -= (alpha * (1/n * 2 * np.dot((y_pred_tmp - y_tmp), X_tmp.T)))
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W
```

 # 1. Постройте график зависимости весов всех признаков от lambda в L2-регуляризации (на данных из урока).

```python
def eval_model_reg2(X, y, iterations, alpha=1e-4, lambdas=1e-8):
    W_arr = []
    err_arr = []
    def _W_(lmbd):
        np.random.seed(42)
        W = np.random.randn(X.shape[0])
        n = X.shape[1]
        for i in range(1, iterations+1):
            y_pred = np.dot(W, X)
            err = calc_mse(y, y_pred)
            W -= alpha * (1/n * 2 * np.dot((y_pred - y), X.T) + lmbd * W) 
        return W, err
    
    if type(lambdas)!=np.ndarray:
        return _W_(lambdas)
    else:
        for lmbd in lambdas:
            W, err = _W_(lmbd)
            W_arr.append(W)
            err_arr.append(err)
        return W_arr, err_arr
        
```

```python
_lambdas = np.arange(1e-4, 1+2e-4, 2e-4)
len(_lambdas)
```

```python
W_arr, err_arr = eval_model_reg2(X_st, y, iterations=4000, alpha=1e-2, lambdas=_lambdas)
W_arr, err_arr = np.array(W_arr), np.array(err_arr)
```

```python
len(err_arr)
```

```python
type(W_arr)
```

```python
w0, w1, w2, w3 = zip(*W_arr)
```

```python
fig, axs = plt.subplots(5, 1, figsize=[10,50])
axs[0].plot(_lambdas, w0)
axs[0].set_title('W0')
axs[1].plot(_lambdas, w1)
axs[1].set_title('W1')
axs[2].plot(_lambdas, w2)
axs[2].set_title('W2')
axs[3].plot(_lambdas, w3)
axs[3].set_title('W3')
axs[4].plot(_lambdas, err_arr)
axs[4].set_title('MSE')
```

# 2. Можно ли к одному и тому же признаку применить сразу и нормализацию, и стандартизацию?

```python
X = np.array([500, 700, 750, 600, 1450, 800, 1500, 2000, 450, 1000])
```

```python
def standart_X(X):
    _mean = X.mean()
    _std = (np.sum((X-_mean)**2)/len(X))**0.5
    standart_X = (X-_mean)/_std
    return standart_X
```

```python
def min_max_X(X):
    _max = X.max()
    _min = X.min()
    min_max_X = (X-_min)/(_max-_min)
    return min_max_X
```

```python
_standart_X = standart_X(X.copy())
_standart_X
```

```python
_min_max_X = min_max_X(X.copy())
_min_max_X
```

```python
_standart_min_max_X = standart_X(_min_max_X)
_standart_min_max_X
```

```python
_min_max_standart_X = min_max_X(_standart_X)
_min_max_standart_X
```

# 3*. Напишите функцию наподобие eval_model_reg2, но для применения L1-регуляризации.

```python
def eval_model_reg1(X, y, iterations, alpha=1e-4, lambdas=1e-8):
    W_arr = []
    err_arr = []
    def _W_(lmbd):
        np.random.seed(42)
        W = np.random.randn(X.shape[0])
        n = X.shape[1]
        for i in range(1, iterations+1):
            y_pred = np.dot(W, X)
            err = calc_mse(y, y_pred)
            W -= alpha * (1/n * 2 * np.dot(X, (y_pred - y)) + lmbd / 2 * W / np.abs(W)) 
        return W, err
    
    if type(lambdas)!=np.ndarray:
        return _W_(lambdas)
    else:
        for lmbd in lambdas:
            W, err = _W_(lmbd)
            W_arr.append(W)
            err_arr.append(err)
        return W_arr, err_arr
        
```
